## Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

#Ans: 5

Pandas provides many functions to manipulate data in a DataFrame. 

- Here are some common functions that you can use:

**1.head()**: Returns the first n rows of the DataFrame. This can be useful to get a quick glimpse of the data.

```python
df = pd.read_csv("services.csv")
df.head(2)
```

**2.tail()**: Returns the last n rows of the DataFrame.

```python
df = pd.read_csv('services.csv')
df.tail(3)
```

**3.describe()**: Generates descriptive statistics for the DataFrame such as count, mean, standard deviation, minimum, and maximum values.

```python
df = pd.read_csv('services.csv')
df.describe()
```

**4.sort_values()**: Sorts the DataFrame by one or more columns.

```python
df = pd.read_csv('services.csv')
df_sorted = df.sort_values(by='eligibility')
df_sorted.head(3)
```

**5.groupby()**: Groups the DataFrame by one or more columns and applies a function to each group. This is often used to calculate aggregate statistics for each group.

```python
df = pd.read_csv('services.csv')
grouped = df.groupby(['location_id']).mean()
print(grouped)
```

**6.drop()**: Drops one or more rows or columns from the DataFrame.

```python
df = pd.read_csv('services.csv')
df_dropped = df.drop(['wait_time'], axis=1)
df_dropped.head()
```

## Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

# Ans: 6
Both Series and DataFrame are mutable in nature, whereas Panel is considered immutable as it is deprecated since version 0.25.0 and replaced with other data structures such as MultiIndex or xarray.

In Pandas, mutable objects can be modified in place, which means the changes made to the object will be reflected in the original object itself. For example, you can add, delete, or modify elements in a Series or a DataFrame after it has been created. However, it is important to note that modifying a mutable object in place can have unintended consequences and it is often safer to create a copy of the object before modifying it.
